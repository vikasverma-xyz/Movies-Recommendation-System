```python
import pandas as pd

movies = pd.read_csv('Movies_data.csv')
```

```python
movies = movies[['id', 'title', 'original_language', 'popularity']]
```

```python
movies
```

```python
movies['tags'] = movies['title' ] + movies['original_language']
```

```python
movies
```

```python
new_data = movies.drop(columns = ['original_language', 'popularity'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=20000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[0])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index = new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
